```python
# read file
file    = kicks[200]
sig, fs = librosa.core.load(file, sr=48000, duration=1)

# process
abs_spectrogram = np.abs(librosa.core.spectrum.stft(sig))
audio_signal = librosa.core.spectrum.griffinlim(abs_spectrogram)

#print(audio_signal, audio_signal.shape)

# write output
#librosa.output.write_wav('test2.wav', audio_signal, fs)
 
```

```python
librosa.display.specshow(abs_spectrogram)
```

```python
log_spec = librosa.amplitude_to_db(np.abs(librosa.core.spectrum.stft(sig)))
librosa.display.specshow(log_spec)
```

```python
IPython.display.Audio(file, rate = sr)
```

```python
IPython.display.Audio(audio_signal, rate = sr)
```

```python
hop_length = 600 # 0.0125*48000
win_length = 2400 # 0.05 * 48000
mel = librosa.feature.melspectrogram(sig, sr=sr, n_fft=2400, hop_length=hop_length, win_length=win_length)
librosa.display.specshow(mel)
```

```python
audio2 = librosa.feature.inverse.mel_to_audio(mel, sr=sr, n_fft=2400, hop_length=hop_length, win_length=win_length)
IPython.display.Audio(audio2, rate = sr)
```
